```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, mean_absolute_percentage_error
from sklearn.model_selection import GridSearchCV
```

```python
diesel_df['Date'] = pd.to_datetime(diesel_df['Date'])
diesel_df.index = diesel_df['Date']
diesel_df.index.freq='MS'
```

```python
#data.to_csv('diesel_df', index=False)
diesel_df[['Original_Diesel_Prices', 'Original_Quantity']] = diesel_df[['Diesel_Prices', 'Quantity']]
```

```python
# Initialize MinMaxScaler
scaler = MinMaxScaler()
diesel_df[['Diesel_Prices', 'Quantity']] = scaler.fit_transform(diesel_df[['Diesel_Prices', 'Quantity']])
```

```python
diesel_df[f'Diesel_Prices_Lag_{1}'] = diesel_df['Diesel_Prices'].shift(1)
diesel_df[f'Diesel_Prices_Lag_{2}'] = diesel_df['Diesel_Prices'].shift(2)
```

```python
diesel_df
```

```python
# Drop rows with NaN values which are the result of lagging
diesel_df.dropna(inplace=True)
```

```python
# Prepare the input and output arrays
feature_columns = ['Diesel_Prices', 'Diesel_Prices_Lag_1', 'Diesel_Prices_Lag_2']
X = diesel_df[feature_columns].values
y = diesel_df['Quantity'].values
```

```python
# Create and train the RandomForest model
rf_model = RandomForestRegressor(max_depth = None, max_features = 'auto', min_samples_leaf = 4, min_samples_split = 5, n_estimators = 100)
rf_model.fit(X_train, y_train)
```

```python
# Making predictions
predictions = rf_model.predict(X_test)
```

```python
# Rescale predictions to original scale
predictions_rescaled = scaler.inverse_transform(np.column_stack([np.zeros(len(predictions)), predictions]))[:, 1]
y_test_rescaled = scaler.inverse_transform(np.column_stack([np.zeros(len(y_test)), y_test]))[:, 1]
```

```python
# Evaluate the model
mse = mean_squared_error(y_test_rescaled, predictions_rescaled)
mae = mean_absolute_error(y_test_rescaled, predictions_rescaled)
rmse = np.sqrt(mse)
r2 = r2_score(y_test_rescaled, predictions_rescaled)
mape = mean_absolute_percentage_error(y_test_rescaled, predictions_rescaled)
```

```python
print("Mean Absolute Error (MAE):", mae)
print("Root Mean Squared Error (RMSE):", rmse)
print("Mean Absolute Percentage Error (MAPE):", mape * 100)
```

```python
results_df = pd.DataFrame({
    'Date': test_dates,
    'Diesel_Prices': test_petrol_prices,
    'Actual_Quantity': y_test_rescaled,
    'Predicted_Quantity': predictions_rescaled
})
```
