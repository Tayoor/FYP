```python
# Load the Excel file
Fuel_Prices = pd.read_excel('Prices.xlsx')
# Display the content of the Excel file
Fuel_Prices.head()
```

```python
# Standardize the date format and extract the year and month for grouping
Fuel_Prices['Date'] = pd.to_datetime(Fuel_Prices['Date'], errors='coerce')
Fuel_Prices['Year'] = Fuel_Prices['Date'].dt.year
Fuel_Prices['Month'] = Fuel_Prices['Date'].dt.month

# Group by year and month and calculate the average prices
Monthly_Fuel_Prices = Fuel_Prices.groupby(['Year', 'Month']).mean().reset_index()

Monthly_Fuel_Prices = Monthly_Fuel_Prices[['Year', 'Month', 'Petrol_Prices', 'Diesel_Prices']]
Monthly_Fuel_Prices
```

```python
diesel_df = pd.merge(monthly_diesel_data, Monthly_Fuel_Prices[['Year', 'Month', 'Diesel_Prices']], on=['Year', 'Month'], how='left')
diesel_df = diesel_df[['Date', 'Diesel_Prices', 'Fuel_Type', 'Quantity']]
diesel_df
```

```python
petrol_df = pd.merge(monthly_petrol_data, Monthly_Fuel_Prices[['Year', 'Month', 'Petrol_Prices']], on=['Year', 'Month'], how='left')
petrol_df = petrol_df[['Date', 'Petrol_Prices', 'Fuel_Type', 'Quantity']]
petrol_df
```

```python
from statsmodels.tsa.stattools import adfuller
```

```python
test_result=adfuller(petrol_df['Quantity'])
```

```python
#Ho: It is non stationary
#H1: It is stationary

def adfuller_test(Quantity):
    result=adfuller(Quantity)
    labels = ['ADF Test Statistic','p-value','#Lags Used','Number of Observations Used']
    for value,label in zip(result,labels):
        print(label+' : '+str(value) )
    if result[1] <= 0.05:
        print("strong evidence against the null hypothesis(Ho), reject the null hypothesis. Data has no unit root and is stationary")
    else:
        print("weak evidence against null hypothesis, time series has a unit root, indicating it is non-stationary ")
    
```

```python
#model = ARIMA(diesel_df["Quantity_Prices"], order=(4, 0, 4), exog=exog_data)
model=ARIMA(petrol_df["Quantity"],order=(2,0,2))
results = model.fit()
```

```python
results1 = results.fittedvalues
results1
```

```python
petrol_df["Quantity"] 
```

```python
results.summary()
```

```python
residuals = petrol_df["Quantity"] - results1
```

```python
plt.figure(figsize=(12, 6))
plt.plot(residuals)
plt.title('Residuals Plot')
plt.show()
```

```python
last_dates = petrol_df['Date'][-5:]
last_actual = petrol_df['Quantity'][-5:]
last_predicted = results1[-5:]
last_predicted
```

```python
# Evaluate the model
mse = mean_squared_error(last_actual, last_predicted)
mae = mean_absolute_error(last_actual, last_predicted)
rmse = np.sqrt(mse)
r2 = r2_score(last_actual, last_predicted)
mape = mean_absolute_percentage_error(last_actual, last_predicted)
```

```python
print("Mean Absolute Error (MAE):", mae)
print("Root Mean Squared Error (RMSE):", rmse)
print("Mean Absolute Percentage Error (MAPE):", mape * 100)
# Print AIC and BIC
print(f"AIC: {results.aic}")
print(f"BIC: {results.bic}")
```

```python
# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(last_dates, last_actual, label='Actual Quantity', marker='o', linestyle='-')
plt.plot(last_dates, last_predicted, label='Predicted Quantity', marker='o', linestyle='-')

plt.title('Comparison of Actual and Predicted Petrol Quantities')
plt.xlabel('Date')
plt.ylabel('Quantity of Petrol')
plt.legend()
plt.grid(True)
# Setting y-axis limits
plt.ylim(200, 1200)  # Set y-axis to range from 100 to 2500
# Rotate date labels for better readability
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```
