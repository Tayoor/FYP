```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, mean_absolute_percentage_error
from sklearn.model_selection import GridSearchCV, train_test_split
```

```python
petrol_df['Date'] = pd.to_datetime(petrol_df['Date'])
petrol_df.index = petrol_df['Date']
petrol_df.index.freq='MS'
```

```python
#data.to_csv('diesel_df', index=False)
petrol_df[['Original_Petrol_Prices', 'Original_Quantity']] = petrol_df[['Petrol_Prices', 'Quantity']]
```

```python
# Initialize MinMaxScaler
scaler = MinMaxScaler()
petrol_df[['Petrol_Prices', 'Quantity']] = scaler.fit_transform(petrol_df[['Petrol_Prices', 'Quantity']])
```

```python
petrol_df[f'Petrol_Prices_Lag_{1}'] = petrol_df['Petrol_Prices'].shift(1)
petrol_df[f'Petrol_Prices_Lag_{2}'] = petrol_df['Petrol_Prices'].shift(2)
```

```python
petrol_df
```

```python
# Drop rows with NaN values which are the result of lagging
petrol_df.dropna(inplace=True)
```

```python
# Prepare the input and output arrays
feature_columns = ['Petrol_Prices', 'Petrol_Prices_Lag_1', 'Petrol_Prices_Lag_2']
X = petrol_df[feature_columns].values
y = petrol_df['Quantity'].values
```

```python
# Create and train the SVR model
svr_model = SVR(kernel='rbf', C=3, gamma=0.07, epsilon=.13)
svr_model.fit(X_train, y_train)
```

```python
# Making predictions
predictions = svr_model.predict(X_test)
```

```python
# Rescale predictions to original scale
predictions_rescaled = scaler.inverse_transform(np.column_stack([np.zeros(len(predictions)), predictions]))[:, 1]
y_test_rescaled = scaler.inverse_transform(np.column_stack([np.zeros(len(y_test)), y_test]))[:, 1]
```

```python
# Evaluate the model
mse = mean_squared_error(y_test_rescaled, predictions_rescaled)
mae = mean_absolute_error(y_test_rescaled, predictions_rescaled)
rmse = np.sqrt(mse)
r2 = r2_score(y_test_rescaled, predictions_rescaled)
mape = mean_absolute_percentage_error(y_test_rescaled, predictions_rescaled)
```

```python
print("Mean Absolute Error (MAE):", mae)
print("Root Mean Squared Error (RMSE):", rmse)
print("Mean Absolute Percentage Error (MAPE):", mape)
```

```python
results_df = pd.DataFrame({
    'Date': test_dates,
    'Petrol_Prices': test_petrol_prices,
    'Actual_Quantity': y_test_rescaled,
    'Predicted_Quantity': predictions_rescaled
})
```
